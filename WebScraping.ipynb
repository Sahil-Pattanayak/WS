#### Using yfinance to Extract Tesla Stock Data

```python
import yfinance as yf

ticker_symbol = "TSLA"
tesla_data = yf.download(ticker_symbol, start="2022-01-01", end="2023-01-01")

tesla_data = tesla_data.reset_index()
tesla_data.to_csv("tesla_stock_data.csv", index=False)

print(tesla_data.head())
```

#### Using yfinance to Extract GME Stock Data

```python
import yfinance as yf

ticker_symbol = "GME"
gme_data = yf.download(ticker_symbol, period="1y")
gme_data.reset_index(inplace=True)
gme_data.to_csv("gme_stock_data.csv", index=False)

print(gme_data.head())
```

####  Ploting Tesla Stock Graph

```python
import yfinance as yf
import matplotlib.pyplot as plt


ticker_symbol = "TSLA"
tesla_data = yf.download(ticker_symbol, period="1y")


def make_graph(data, title):
    plt.figure(figsize=(12, 6))
    plt.plot(data.index, data['Close'], label="Close Price")
    plt.title(title)
    plt.xlabel("Date")
    plt.ylabel("Stock Price")
    plt.legend()
    plt.grid(True)
    plt.show()


graph_title = "Tesla Stock Price Over the Last Year"
make_graph(tesla_data, graph_title)
```
